```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors
a = np.array([2, 3])
b = np.array([3, -2])
c = a + b  # Vector addition

# Setup the plot
plt.figure(figsize=(8, 8))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Plot vector a
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a')

# Plot vector b starting from head of vector a
plt.quiver(a[0], a[1], b[0], b[1], angles='xy', scale_units='xy', scale=1, color='green', label='Vector b')

# Plot resultant vector c from origin
plt.quiver(0, 0, c[0], c[1], angles='xy', scale_units='xy', scale=1, color='red', label='Result a + b')

# Annotations
plt.text(2, 3.2, 'a', color='blue', fontsize=12)
plt.text(5.2, 1, 'c = a + b', color='red', fontsize=12)
plt.text(5, 1.2, f'({c[0]}, {c[1]})', color='black', fontsize=10)

# Set the grid and limits
plt.xlim(-1, 7)
plt.ylim(-3, 5)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Vector Addition Visualization")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
a = np.array([2, 3])
b = np.array([3, -2])
c = a - b  # Subtraction

# Setup plot
plt.figure(figsize=(8, 8))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Plot vector a
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a')

# Plot vector b
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='green', label='Vector b')

# Plot result c = a - b
plt.quiver(0, 0, c[0], c[1], angles='xy', scale_units='xy', scale=1, color='red', label='a - b (result)')

# Annotations
plt.text(a[0], a[1]+0.3, 'a', color='blue', fontsize=12)
plt.text(b[0], b[1]-0.3, 'b', color='green', fontsize=12)
plt.text(c[0]-0.8, c[1]+0.3, 'a - b', color='red', fontsize=12)

# Set the grid and limits
plt.xlim(-3, 5)
plt.ylim(-3, 6)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Vector Subtraction Visualization")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Original vector
v = np.array([2, 3])

# Scaled vectors
v_2 = 2 * v
v_neg_half = -0.5 * v

# Plot setup
plt.figure(figsize=(10, 6))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Plot original
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vector')

# Plot scaled (2)
plt.quiver(0, 0, v_2[0], v_2[1], angles='xy', scale_units='xy', scale=1, color='green', label='2  Vector')

# Plot scaled (-0.5)
plt.quiver(0, 0, v_neg_half[0], v_neg_half[1], angles='xy', scale_units='xy', scale=1, color='red', label='-0.5  Vector')

# Labels
plt.text(v[0], v[1]+0.3, 'v', color='blue')
plt.text(v_2[0], v_2[1]+0.3, '2v', color='green')
plt.text(v_neg_half[0]-0.5, v_neg_half[1], '-0.5v', color='red')

plt.xlim(-2, 5)
plt.ylim(-2, 7)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.title("Multiplying a Vector by a Scalar")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define two vectors
a = np.array([2, 3])
b = np.array([4, 1])

# Dot product
dot_product = np.dot(a, b)

# Magnitudes
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)

# Cosine of angle
cos_theta = dot_product / (norm_a * norm_b)
angle_rad = np.arccos(cos_theta)  # in radians
angle_deg = np.degrees(angle_rad)  # in degrees

print(f"Angle between vectors: {angle_deg:.2f}")

# Plot
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='green', label='Vector b')
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.gca().set_aspect('equal')
plt.grid(True)
plt.title(f"Angle = {angle_deg:.2f}")
plt.legend()
plt.show()
```

```python
import numpy as np

# Define word embedding vectors (fake example)
king = np.array([0.8, 0.6])
queen = np.array([0.7, 0.7])
dog = np.array([0.2, 0.9])

# Dot product similarity
def dot_similarity(a, b):
    return np.dot(a, b)

print("king  queen:", dot_similarity(king, queen))  # high
print("king  dog:", dot_similarity(king, dog))      # lower
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define 3D vectors
a = np.array([2, 3, 0])
b = np.array([4, 4, 0])
cross_ab = np.cross(a, b)

print("Cross Product (a  b):", cross_ab)

# 3D Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
origin = [0, 0, 0]

# Vectors
ax.quiver(*origin, *a, color='blue', label='a')
ax.quiver(*origin, *b, color='green', label='b')
ax.quiver(*origin, *cross_ab, color='red', label='a  b')

ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])
ax.set_title("Cross Product of Vectors")
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
a = np.array([3, 4])
v = np.array([4, 0])

# Calculate projection of a onto v
proj = (np.dot(a, v) / np.dot(v, v)) * v

# Plot
plt.quiver(0, 0, *a, color='blue', label='a', angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, *v, color='red', label='v', angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, *proj, color='orange', label='proj_v a', angles='xy', scale_units='xy', scale=1)

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid()
plt.gca().set_aspect('equal')
plt.legend()
plt.title("Projection of a onto v")
plt.show()
```
