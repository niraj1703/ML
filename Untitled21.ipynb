# Project 1: House Price Prediction (Polynomial Regression)
__ðŸ”¹ Problem__
__Predict house prices based on their size (sq. ft.).__

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# âœ… Step 1: Create synthetic data (Size in sq. ft vs Price in $1000)
X = np.array([500, 750, 1000, 1250, 1500, 1750, 2000, 2250, 2500]).reshape(-1, 1)
Y = np.array([70, 110, 150, 200, 260, 330, 410, 500, 600])

# âœ… Step 2: Transform data to polynomial features (degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# âœ… Step 3: Fit polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# âœ… Step 4: Predict
Y_pred = model.predict(X_poly)

# âœ… Step 5: Print formula
print("Intercept (b0):", model.intercept_)
print("Coefficients (b1, b2):", model.coef_)

# âœ… Step 6: Visualization
plt.scatter(X, Y, color='red', label='Actual Data')
plt.plot(X, Y_pred, color='blue', label='Polynomial Fit')
plt.xlabel("House Size (sq. ft.)")
plt.ylabel("Price (in $1000)")
plt.title("House Price Prediction - Polynomial Regression")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# âœ… Step 1: Sample Data (Rainfall mm, Temperature Â°C â†’ Yield in tons/hectare)
X = np.array([
    [400, 20],
    [420, 22],
    [430, 23],
    [440, 25],
    [460, 28],
    [470, 30],
    [480, 32],
    [500, 33],
    [520, 34],
    [530, 35]
])
y = np.array([1.8, 2.0, 2.4, 3.0, 3.4, 3.3, 3.1, 3.0, 2.7, 2.5])  # Crop yield

# âœ… Step 2: Polynomial features (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# âœ… Step 3: Fit polynomial regression model
model = LinearRegression()
model.fit(X_poly, y)

# âœ… Step 4: Predict yield
y_pred = model.predict(X_poly)

# âœ… Step 5: Print equation coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# âœ… Step 6: Visualization (Rainfall vs Yield for fixed temp)
import pandas as pd
df = pd.DataFrame(X, columns=['Rainfall', 'Temp'])
df['Yield'] = y
df['Predicted_Yield'] = y_pred

plt.scatter(df['Rainfall'], df['Yield'], color='red', label='Actual')
plt.plot(df['Rainfall'], df['Predicted_Yield'], color='blue', label='Polynomial Fit')
plt.xlabel("Rainfall (mm)")
plt.ylabel("Yield (tons/hectare)")
plt.title("Crop Yield Prediction (Fixed Temperature)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

# âœ… Step 1: Create a sample dataset
data = {
    'size_sqft': [1500, 1800, 2400, 3000, 3500, 4000],
    'age_years': [10, 15, 20, 5, 8, 12],
    'bedrooms': [3, 4, 3, 5, 4, 4],
    'price': [250000, 310000, 400000, 500000, 550000, 600000]
}
df = pd.DataFrame(data)

# âœ… Step 2: Independent and target variables
X = df[['size_sqft', 'age_years', 'bedrooms']]
y = df['price']

# âœ… Step 3: Polynomial Features (degree=2)
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

# âœ… Step 4: Fit model
model = LinearRegression()
model.fit(X_poly, y)

# âœ… Step 5: Predict
y_pred = model.predict(X_poly)

# âœ… Step 6: Visualize prediction vs actual
plt.plot(range(len(y)), y, 'go-', label='Actual Price')
plt.plot(range(len(y_pred)), y_pred, 'ro--', label='Predicted Price')
plt.xlabel('House Index')
plt.ylabel('Price (â‚¹)')
plt.title('Polynomial Regression: House Price Prediction')
plt.legend()
plt.grid(True)
plt.show()
```
