```python
df['hours_studied'] = np.round(np.random.uniform(1, 10, size=14), 2)
df['exam_score'] = np.round(50 + df['hours_studied'] * 5 + np.random.normal(0, 5, size=14), 2)  # Adding some noise
```

```python
df
```

```python
plt.scatter(df['hours_studied'], df['exam_score'], color='green')
# Trend line
z = np.polyfit(df['hours_studied'], df['exam_score'], 1)
p = np.poly1d(z)
plt.plot(df['hours_studied'], p(df['hours_studied']), color='red')
plt.title("Study Hours vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.grid()
plt.show()

```
