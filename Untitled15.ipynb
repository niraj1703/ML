```python
import numpy as np
import matplotlib.pyplot as plt

# Define two vectors
v1 = np.array([2, 3])
v2 = np.array([4, 1])

# Vector addition
v_sum = v1 + v2
dot_product = np.dot(v1, v2)

# Plot
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green', label='v2')
plt.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='red', label='v1 + v2')

plt.xlim(0, 7)
plt.ylim(0, 7)
plt.grid()
plt.legend()
plt.title(f"Vector Addition | Dot Product = {dot_product}")
plt.show()
```

```python
import seaborn as sns

# Sample data
data = [3, 7, 8, 10, 6, 4, 5, 7, 9, 6]
mean = np.mean(data)
var = np.var(data)

# Plot
sns.histplot(data, kde=True)
plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
plt.title(f"Distribution (Variance = {var:.2f})")
plt.legend()
plt.show()

```

```python
# Bayes visualization using prior and likelihood
P_disease = 0.01
P_positive_given_disease = 0.95
P_positive_given_no_disease = 0.05
P_no_disease = 1 - P_disease

numerator = P_positive_given_disease * P_disease
denominator = numerator + (P_positive_given_no_disease * P_no_disease)
posterior = numerator / denominator

labels = ['Has Disease', 'Does Not Have Disease']
sizes = [posterior, 1 - posterior]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Bayes Posterior Probability (After Positive Test)")
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
y = x**2

plt.plot(x, y, label='f(x) = x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent Path')

# simulate steps
x0 = 8
learning_rate = 0.3
x_vals = [x0]
for _ in range(10):
    grad = 2 * x0  # derivative of x is 2x
    x0 = x0 - learning_rate * grad
    x_vals.append(x0)

y_vals = [val**2 for val in x_vals]
plt.plot(x_vals, y_vals, 'ro--', label='Descent Steps')
plt.legend()
plt.grid()
plt.show()

```
