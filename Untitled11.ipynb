```python
from sklearn.datasets import load_iris
from lightgbm import Dataset

data = load_iris()
X = data['data']
y = data['target']

lgb_dataset = Dataset(X, label=y)
```

```python
params = {'objective': 'multiclass', 'num_class': 3, 'metric': 'multi_logloss'}
model = lgb.train(params, lgb_dataset, num_boost_round=50)
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import lightgbm as lgb

X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

train_data = lgb.Dataset(X_train, label=y_train)
params = {'objective': 'binary', 'metric': 'binary_logloss'}
model = lgb.train(params, train_data, num_boost_round=100)
```

```python
params = {
    'objective': 'binary',
    'metric': 'binary_logloss',
    'num_leaves': 31,
    'max_depth': 6
}

```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
y_pred_class = (y_pred > 0.5).astype(int)
accuracy_score(y_test, y_pred_class)

```

```python
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMClassifier

model = LGBMClassifier()
params = {
    'num_leaves': [31, 50],
    'learning_rate': [0.1, 0.01]
}
grid = GridSearchCV(model, params, cv=3)
grid.fit(X_train, y_train)
print(grid.best_params_)

```

```python
import matplotlib.pyplot as plt
lgb.plot_importance(model)
plt.show()
```

```python
valid_data = lgb.Dataset(X_test, label=y_test)
model = lgb.train(params, train_data, valid_sets=[valid_data], early_stopping_rounds=10)
```

```python
model.save_model('model.txt')
model = lgb.Booster(model_file='model.txt')
```

```python
import lightgbm as lgb
import pandas as pd

data = load_iris()
X, y = data.data, data.target
train_data = lgb.Dataset(X, label=y)

print(type(train_data))
```

```python
params = {
    'objective': 'multiclass',
    'num_class': 3,
    'metric': 'multi_logloss'
}

model = lgb.train(params, train_data, num_boost_round=10)
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
clf = LGBMClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))

```

```python
cv_results = lgb.cv(params, train_data, nfold=5, num_boost_round=50, early_stopping_rounds=10)
print(cv_results['multi_logloss-mean'][-1])
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'num_leaves': [15, 31, 63],
    'learning_rate': [0.01, 0.1],
}
grid = GridSearchCV(LGBMClassifier(), param_grid=params, cv=3)
grid.fit(X_train, y_train)
print(grid.best_params_)


```

```python
clf = LGBMClassifier(n_estimators=100)
clf.fit(X_train, y_train,
        eval_set=[(X_test, y_test)],
        early_stopping_rounds=10,
        verbose=False)
```

```python
import matplotlib.pyplot as plt
lgb.plot_importance(clf)
plt.show()

```
