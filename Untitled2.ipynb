```python
result, error = integrate.quad(lambda x: x**2, 0, 1)
print(f"Integral result: {result}, Error: {error}")
```

```python
dbl_result, _ = integrate.dblquad(lambda y, x: x*y, 0, 1, lambda x: 0, lambda x: 1)
print(f"Double integral: {dbl_result}")
```

```python
a = [12, 15, 14, 10, 13]
b = [13, 14, 15, 15, 14]
t_stat, p_val = stats.ttest_ind(a, b)
print(f"T-stat: {t_stat}, P-value: {p_val}")
```

```python
t = np.linspace(0, 1, 100, False)
x = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * 20 * t)
# Filter design
b, a = butter(4, 0.1)
y = lfilter(b, a, x)
print(y[:5])  # filtered output
```
