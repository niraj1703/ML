```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
```

```python
df = sns.load_dataset('tips')
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title('Total Bill vs Tip')
plt.show()
```

```python
df['order'] = df.index
sns.lineplot(x='order', y='total_bill', data=df)
plt.title('Total Bill Over Orders')
plt.show()
```

```python

sns.barplot(x='day', y='tip', data=df, estimator='mean')
plt.title('Average Tip per Day')
plt.show()
```

```python
sns.histplot(df['total_bill'], bins=20, kde=True)
plt.title('Distribution of Total Bill')
plt.show()
```

```python
sns.boxplot(x='sex', y='total_bill', data=df)
plt.title('Total Bill by Gender')
plt.show()
```

```python
sns.countplot(x='sex', data=df)
plt.title('Gender Count')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', data=df, split=True)
plt.title('Total Bill Distribution by Day and Gender')
plt.show()
```

```python
sns.swarmplot(x='sex', y='tip', data=df)
plt.title('Tip Distribution by Gender')
plt.show()
```

```python
sns.catplot(x='day', y='tip', hue='smoker', kind='box', data=df)
plt.title('Tips by Day (Smoker vs Non-Smoker)')
plt.show()
```

```python
g = sns.FacetGrid(df, col='sex')
g.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=df)
plt.title('Linear Regression: Total Bill vs Tip')
plt.show()
```

```python
flights = sns.load_dataset('flights')
pivot = pivot = flights.pivot(index='month', columns='year', values='passengers')
sns.heatmap(pivot, cmap='coolwarm', annot=True, fmt="d")
plt.title('Monthly Passengers Heatmap')
plt.show()
```

```python
sns.clustermap(pivot, cmap='viridis', annot=True)
plt.title('Clustered Heatmap')
plt.show()
```

```python
print(sns.get_dataset_names())
tips = sns.load_dataset("tips")
penguins = sns.load_dataset("penguins")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
flights = sns.load_dataset("flights")
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Total Bill vs Tip")
plt.show()

sns.barplot(x="day", y="tip", data=tips, estimator='mean')
plt.title("Average Tip per Day")
plt.show()

sns.countplot(x="sex", data=tips)
plt.title("Gender Distribution")
plt.show()

sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title("Total Bill Distribution")
plt.show()
```

```python
penguins = penguins.dropna()
sns.boxplot(x="species", y="body_mass_g", data=penguins)
plt.title("Body Mass by Species")
plt.show()

sns.violinplot(x="island", y="flipper_length_mm", data=penguins)
plt.title("Flipper Length by Island")
plt.show()

sns.pairplot(penguins, hue="species")
plt.show()
```

```python
sns.countplot(x="sex", data=titanic)
plt.title("Passenger Gender Count")
plt.show()

sns.catplot(x="pclass", hue="survived", kind="count", data=titanic)
plt.title("Survival by Class")
plt.show()

sns.violinplot(x="class", y="age", hue="survived", data=titanic, split=True)
plt.title("Age Distribution by Class and Survival")
plt.show()
```

```python
fig, ax = plt.subplots()
sns.boxplot(x="day", y="tip", data=tips, ax=ax)
ax.set_title("Tip by Day")
ax.set_ylabel("Tip Amount")
ax.set_xlabel("Day of Week")
plt.savefig("tip_by_day.png")
plt.show()
```

```python
grouped = tips.groupby("day")["tip"].mean().reset_index()
sns.barplot(x="day", y="tip", data=grouped)
plt.title("Mean Tip by Day")
plt.show()

pivot2 = tips.pivot_table(index="day", columns="sex", values="tip", aggfunc="mean")
sns.heatmap(pivot2, annot=True, cmap="YlGnBu")
plt.title("Average Tip by Day and Gender")
plt.show()

merged = pd.concat([tips, pd.get_dummies(tips["sex"])], axis=1)
sns.scatterplot(x="total_bill", y="tip", hue="Female", data=merged)
plt.title("Bill vs Tip Colored by Female")
plt.show()

```
