```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame()

num_rows = 3
```

```python
df['Day'] = np.random.choice(
    ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'],
    size=num_rows
)
```

```python
df['Temperature'] = np.random.randint(20, 40, size=num_rows)
```

```python
sample = df.head(7)

plt.plot(sample['Day'], sample['Temperature'], color='blue')
plt.title("Sample Daily Temperature")
plt.xlabel("Day")
plt.ylabel("Temperature (Â°C)")
plt.grid()
plt.show()
```

```python
df['Month'] = np.random.randint(1, 30, )
df['Sales'] = np.random.choice([60000, 10000000],)
```

```python
plt.bar(df['Month'], df['Sales'], color='orange')
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
df['Age'] = np.random.randint(1, 30, )
```

```python
plt.hist(df['Age'], bins=10, color='green', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
df['Hours'] = np.random.randint(1, 10, )
df['Score'] = np.random.randint(1, 30, )
```

```python
plt.scatter(df['Hours'], df['Score'], color='red')
plt.title("Study Hours vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.grid()
plt.show()
```
