```python
iris = load_iris()
X, y = iris.data, iris.target
dtrain = xgb.DMatrix(X, label=y)

print(dtrain.num_row(), dtrain.num_col())
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

model = XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
params = {'max_depth': 3, 'eta': 0.1, 'objective': 'multi:softmax', 'num_class': 3}
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

model = xgb.train(params, dtrain, num_boost_round=10)
preds = model.predict(dtest)

from sklearn.metrics import classification_report
print(classification_report(y_test, preds))
```

```python
cv_results = xgb.cv(params, dtrain, num_boost_round=10, nfold=3, metrics="mlogloss", as_pandas=True)
print(cv_results)
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'max_depth': [2, 4, 6],
    'learning_rate': [0.01, 0.1],
}
model = XGBClassifier()
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv=3)
grid.fit(X_train, y_train)

print("Best parameters:", grid.best_params_)
```

```python
from xgboost import XGBClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = XGBClassifier(n_estimators=100, eval_metric='mlogloss')
model.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    verbose=False
)

```

```python
from sklearn.datasets import load_iris
import xgboost as xgb

iris = load_iris()
X, y = iris.data, iris.target
dtrain = xgb.DMatrix(X, label=y)

print(dtrain.num_row(), dtrain.num_col())
```

```python
params = {'max_depth': 3, 'eta': 0.1, 'objective': 'multi:softmax', 'num_class': 3}
model = xgb.train(params, dtrain, num_boost_round=10)

preds = model.predict(dtrain)
print(preds[:5])
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
cv_results = xgb.cv(params, dtrain, num_boost_round=10, nfold=5,
                    metrics='mlogloss', as_pandas=True)
print(cv_results)
```

```python
from sklearn.model_selection import GridSearchCV

params = {'max_depth': [2, 4, 6], 'learning_rate': [0.01, 0.1]}
grid = GridSearchCV(XGBClassifier(), param_grid=params, cv=3)
grid.fit(X_train, y_train)

print("Best parameters:", grid.best_params_)
```

```python
model = XGBClassifier(n_estimators=100)
model.fit(X_train, y_train, early_stopping_rounds=5,
          eval_set=[(X_test, y_test)], verbose=False)
```

```python
import matplotlib.pyplot as plt
xgb.plot_importance(model)
plt.show()
```

```python
explainer = shap.Explainer(model)
shap_values = explainer(X_test)

shap.plots.waterfall(shap_values[0])
```
