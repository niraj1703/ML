```python
import matplotlib.pyplot as plt 
fig = plt.figure(figsize=(4, 2), 
                 facecolor='green',      # facecolor is a graph outside color 
                 layout='constrained')   # set equaly space in graph
fig.suptitle('A nice Matplotlib Figure') # title for a picture like heading for a picture
ax = fig.add_subplot()
ax.set_title('Axes', 
             loc='left',           # here text loc = place word in left side 
             fontstyle='oblique',  # fontstyle = create text in oblique
             fontsize='medium')    #fontsize = size of the text              
```

```python
fig = plt.figure(figsize=(2, 2), facecolor='lightskyblue',
                 layout='constrained')
fig.suptitle('Figure')
ax = fig.add_subplot()
ax.set_title('Axes', loc='left', fontstyle='oblique', fontsize='medium')
```
