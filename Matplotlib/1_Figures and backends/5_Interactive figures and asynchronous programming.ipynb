>Event loops

```python
exec_count = 0   # A counter to keep track of how many commands you have run.
while True:      # Infinite loop — keeps running until you manually stop it (with Ctrl+C or by closing the program).

    inp = input(f"[{exec_count}] > ")        # Read    Displays a prompt like:
    
    ret = eval(inp)                          # Evaluate    eval() evaluates the string as Python code.
                                             #"2 + 3" → becomes actual Python code → result = 5.
                                             #⚠ Important: eval() works only for expressions (not full statements like for, while, def).
    
    print(ret)                               # Print  Prints the evaluated result.
    exec_count += 1                          # Loop   Increments the counter, so the next prompt becomes:
```

>Blocking functions

```python
fig, ax = plt.subplots()  # Creates a new figure and one Axes object.


ln, = ax.plot([], [])   #  Creates an empty line object (Line2D) on the plot.
                        #The comma (ln, = ...) unpacks the single returned object into ln.
                        #(Because ax.plot() returns a list of line objects).

while True:             #Infinite loop → keeps running continuously.

    x, y = get_new_data()   # You are expected to define a function get_new_data() that returns new X and Y values each time the loop runs.


    ln.set_data(x, y)  # Updates the existing line with new X & Y data, instead of creating a new line each time. Much faster than re-plotting.
    plt.pause(1)    # Pauses for 1 second → allows the figure to refresh/redraw.
```

>Explicitly spinning the event loop

```python
import time
import matplotlib.pyplot as plt
import numpy as np
plt.ion()

fig, ax = plt.subplots()
th = np.linspace(0, 2*np.pi, 512)
ax.set_ylim(-1.5, 1.5)

ln, = ax.plot(th, np.sin(th))

def slow_loop(N, ln):
    for j in range(N):
        time.sleep(.1)  # to simulate some work
        ln.figure.canvas.flush_events()

slow_loop(100, ln)
```

```python
def slow_loop(N, ln):
    for j in range(N):
        time.sleep(.1)  # to simulate some work
        if j % 10:
            ln.set_ydata(np.sin(((j // 10) % 5 * th)))
            ln.figure.canvas.draw_idle()

        ln.figure.canvas.flush_events()

slow_loop(100, ln)
```
