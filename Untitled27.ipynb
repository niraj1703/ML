Gradient Descent is an optimized Algorithms used for minimizing the loss function in various machine learning algorithms .

$b = b - \alpha \cdot db$  

**Learning Rate:**  
Learning rate ($\alpha$) is a tuning parameter in an optimization algorithm that determines the step size at each iteration while moving toward a minimum of a loss function.
